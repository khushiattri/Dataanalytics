```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ecommerce_product_dataset.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(30,10))
by=sns.countplot(data = df, x="Discount")
by.bar_label(by.containers[0])
plt.title("Discount Distribution")
plt.show()
```

#from the above chart we have analyzed the count of various product categories offering discount rate to their services. 
Highest being in the range of 0.41 with a count of 32 . 

```python
gb= df.groupby("Category").agg({"Price":'mean', "Sales": 'mean'})
print(gb)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(gb, annot=True)
plt.show()
```

#from the above chart we have concluded that the increase in price of various category products habe an inverse relationship to their corresponding sales.

```python
gb1= df.groupby("Rating").agg({"Price":'mean', "Sales": 'mean'})
print(gb1)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(gb1, annot=True)
plt.show()
```

#from the above chart we have concluded that the ratings of the products have a signifucant impact on their sales and prices.

Distribution on the basis of Cities

```python
print(df["City"].unique())
```

```python
Austin = df.loc[(df['City'] == "Austin")].count()
Queens = df.loc[(df['City'] == "Queens")].count()
Houston = df.loc[(df['City'] == "Houston")].count()
Dayton = df.loc[(df['City'] == "Dayton")].count()
Brooklyn = df.loc[(df['City'] == "Brooklyn")].count()
p = ["Austin","Queens","Houston","Dayton","Brooklyn"]
mlist= [Austin["Sales"],Queens["Sales"],Houston["Sales"],Dayton["Sales"],Brooklyn["Sales"]]
plt.pie(mlist,labels= p)
plt.title("Distribution on the basis of different Cities")
plt.show()
```
